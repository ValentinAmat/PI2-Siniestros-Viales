```python
import pandas as pd
```

<h3>Fusionamos los archivos "hechos_homicidios.csv" y "victimas_homicidios.csv":</h3>

```python
# Cargamos el archivo de hechos:
hechos_df = pd.read_csv(r"../Archivos producidos/hechos_homicidios.csv")
hechos_df.head()
```

```python
# Cargamos el archivo de víctimas:
victimas_df = pd.read_csv(r"../Archivos producidos/victimas_homicidios.csv")
victimas_df.head()
```

```python
# Unimos las columnas a utilizar de ámbos dataframes:
homicidios_df = pd.merge(hechos_df, victimas_df[["ID_hecho", "ROL", "SEXO", "EDAD"]], left_on="ID", right_on="ID_hecho", how="inner").drop("ID_hecho", axis=1)
homicidios_df.head()
```

```python
# Renombramos las columnas:
homicidios_df["ROL_VICTIMA"] = homicidios_df["ROL"]
homicidios_df["SEXO_VICTIMA"] = homicidios_df["SEXO"]
homicidios_df["EDAD_VICTIMA"] = homicidios_df["EDAD"]
homicidios_df["CALLE"] = homicidios_df["Calle"]
homicidios_df["CRUCE"] = homicidios_df["Cruce"]
homicidios_df["LATITUD"] = homicidios_df["pos x"]
homicidios_df["LONGITUD"] = homicidios_df["pos y"]

homicidios_df.drop(["ROL", "SEXO", "EDAD", "Calle", "Cruce", "pos x", "pos y"], axis=1, inplace=True)

homicidios_df.head(3)
```

```python
# Revisamos las columnas y sus tipos de dato:
homicidios_df.dtypes
```

```python
# Corregimos los tipos de dato de las columnas "FECHA", "EDAD_VICTIMA" y "HH":
homicidios_df["FECHA"] = pd.to_datetime(homicidios_df["FECHA"])
homicidios_df["EDAD_VICTIMA"] = homicidios_df["EDAD_VICTIMA"].astype("Int64")
homicidios_df["HH"] = homicidios_df["HH"].astype("Int64")

print(homicidios_df["FECHA"].dtype)
print(homicidios_df["EDAD_VICTIMA"].dtype)
print(homicidios_df["HH"].dtype)
```

```python
# Creamos una nueva columna llamada "DIA_SEMANA" que posee el día de la semana en que ocurrió el accidente:
homicidios_df["DIA_SEMANA"] = homicidios_df["FECHA"].dt.day_name()
homicidios_df["DIA_SEMANA"].head()
```

```python
# Visualizamos el DataFrame terminado:
homicidios_df.head()
```

```python
# Guardamos el archivo:
homicidios_df.to_csv(r"../Archivos producidos/homicidios_fusion.csv", index=False)
```
